4- A transformada de Fourier é uma ferramenta matemática que tem a função de passar o sinal para o ramo das frequências. Permite analisar sinais de áudio por exemplo em graves e agudos. A - Pela sua forma simplificada fast fourier transform, os sinais podem não ser ideais para a análise por conta de amostragens que não batem com as frequencias fundamentais por conta de estarem fora de fase. B- Esse problema é corrigido com outra função matemática aplicada ao sinal original para amortecer o sistema, processo esse chamado de janelamento.
C- O janelamento causa perdas na amplitude do sinal, que podem ser corrigidor por um fator de correção, analisando a diferença da amplitude do sinal original e o sinal resultante.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import librosa
```

```python
#carregar strings e violins
string, sr_string = librosa.load('20String.wav',sr=None)
violin, sr_violin = librosa.load('20Violins .wav',sr=None)
```

```python
def plot_signal(signal, sample_rate, title):
    duration = len(signal) / sample_rate
    t = np.linspace(0, duration, len(signal), endpoint=False)
    plt.figure(figsize=(10, 4))
    plt.plot(t, signal)
    plt.xlabel('Time (seconds)')
    plt.ylabel('Amplitude')
    plt.title(title)
    plt.grid(True)
    plt.show()

plot_signal(string,sr_string,"20Strings")
plot_signal(violin,sr_violin,"20Strings")
```

```python
def plot_fft(signal, sample_rate, title):
    threshold=0.1
    n = len(signal)
    freq = np.fft.fftfreq(n, 1 / sample_rate)
    fft_values = np.abs(np.fft.fft(signal)) / n

    plt.figure(figsize=(10, 4))
    plt.plot(freq[:n // 2], fft_values[:n // 2])
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.title(title)
    plt.grid(True)


plot_fft(string,sr_string,"20Strings")
plot_fft(violin,sr_violin,"20Strings")
```

```python
def plot_spectrogram(sinal, sample_rate, title):
    plt.figure(figsize=(10, 4))
    spectrogram = np.abs(librosa.stft(sinal))
    spectrogram_db = librosa.amplitude_to_db(spectrogram, ref=np.max)
    librosa.display.specshow(spectrogram_db, sr=sample_rate, x_axis='time', y_axis='linear')
    plt.colorbar(format='%2.0f dB')
    plt.title(title)
    plt.tight_layout()
    plt.show()

plot_spectrogram(string, sr_string, "Espectrograma - 20Strings")
plot_spectrogram(violin, sr_violin, "Espectrograma - Violino")
```

```python
#Função para calcular e plotar o mel spectrograma
def plot_mel_spectrogram(sinal, sample_rate, title):
  plt.figure(figsize=(10,4))
  mel_spectrogram = librosa.feature.melspectrogram(y=sinal, sr=sample_rate)
  mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)
  y_axis_values = librosa.mel_frequencies()
  librosa.display.specshow(mel_spectrogram_db, sr=sample_rate, x_axis='time', y_axis='mel', y_coords=y_axis_values)
  plt.colorbar(format='%2.0f dB')
  plt.title(title)
  plt.tight_layout()
  plt.show()

plot_mel_spectrogram(string,sr_string,"20Strings")
plot_mel_spectrogram(violin,sr_violin,"20Strings")
```

```python
def plot_mfcc(sinal, sample_rate, n_mfcc, title):
    plt.figure(figsize=(10, 4))
    mfccs = librosa.feature.mfcc(y=sinal, sr=sample_rate, n_mfcc=n_mfcc)
    librosa.display.specshow(mfccs, x_axis='time', sr=sample_rate)
    plt.colorbar(format='%+2.0f dB')
    plt.title(title)
    plt.tight_layout()
    plt.show()

plot_mfcc(string, sr_string, 13, "MFCC Strings")
plot_mfcc(violin, sr_violin, 13, "MFCC Violins")
```

```python
import librosa
import numpy as np
import librosa.display
import matplotlib.pyplot as plt

def plot_fft_with_max_frequency(signal, sample_rate, title):
    threshold = 0.1
    n = len(signal)
    freq = np.fft.fftfreq(n, 1 / sample_rate)
    fft_values = np.abs(np.fft.fft(signal)) / n

    plt.figure(figsize=(10, 6))
    plt.plot(freq[:n // 2], fft_values[:n // 2])
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.title(title)

    # Encontra a maior frequência no espectro
    max_freq_index = np.argmax(fft_values)
    max_freq = freq[max_freq_index]

    # Plota uma linha vertical indicando a maior frequência
    plt.axvline(x=max_freq, color='red', linestyle='--', label=f'Frequencia com maior amplitude: {max_freq:.2f} Hz')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    plt.axvline(x=4850, color='green', linestyle='--', label=f'Frequencia com maior amplitude: {max_freq:.2f} Hz')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

# Exemplo de uso
plot_fft_with_max_frequency(string, sr_string, "FFT com Maior Frequência - 20Strings")

```

```python
import librosa
import numpy as np
import librosa.display
import matplotlib.pyplot as plt

def resample_and_plot(audio, original_sample_rate, target_sample_rate):
    # Realizar o resample do áudio para a nova taxa de amostragem
    resampled_signal = librosa.resample(audio, orig_sr=original_sample_rate, target_sr=target_sample_rate)

    # Plotar o sinal original e o sinal resample
    plt.figure(figsize=(10, 4))
    plt.plot(np.arange(len(audio)) / original_sample_rate, audio, label='Sinal Original')
    plt.plot(np.arange(len(resampled_signal)) / target_sample_rate, resampled_signal, label='Sinal Resample')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Amplitude')
    plt.title('Comparação entre Sinal Original e Sinal Resample')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Definir as taxas de amostragem desejadas
target_sample_rate_012 = int(4800 * 1.2)
target_sample_rate_100 = int(4800 * 10)

# Chamar a função com os argumentos necessários
resample_012x = resample_and_plot(string, sr_string, target_sample_rate_012)
resample_100x = resample_and_plot(string, sr_string, target_sample_rate_100)
```

```python
def resample_audio(audio, original_sample_rate, target_sample_rate):
    # Realizar o resample do áudio para a nova taxa de amostragem
    resampled_signal = librosa.resample(audio, orig_sr=original_sample_rate, target_sr=target_sample_rate)
    return resampled_signal
```

```python
def apply_window(signal, window_type):
    windows = {
        'bartlett': np.bartlett,
        'blackman': np.blackman,
        'hamming': np.hamming,
        'hanning': np.hanning,
        # Adicione mais tipos de janela aqui, se necessário
    }

    if window_type in windows:
        window_func = windows[window_type]
        window = window_func(len(signal))
        return signal * window
```

Com um resample de 10x a maior frequencia, o audio está totalmente sobreposto sobre o audio original, que não acontece sobre o 1.2x

```python
# Realizar o resample do áudio
resample_012x = resample_audio(string, sr_string, target_sample_rate_012)
resample_100x = resample_audio(string, sr_string, target_sample_rate_100)

# Aplicar janelamento ao sinal resample
janelamento_012 = apply_window(resample_012x, 'bartlett')
```

```python
plot_signal(string,sr_string,"20Strings")
plot_signal(janelamento_012,target_sample_rate_012,"Resampled com janelamento")
```

```python
plot_fft(string,sr_string,"20Strings")
plot_fft(janelamento_012,target_sample_rate_012,"Resampled com janelamento")
```

```python
plot_fft(resample_100x,target_sample_rate_100,"Resampled 10x")
plot_fft(janelamento_012,target_sample_rate_012,"Resampled 1.2x com janelamento")
```

```python
def apply_amplitude_correction(signal, target_amplitude):
    current_amplitude = np.max(np.abs(signal))
    correction_factor = target_amplitude / current_amplitude
    corrected_signal = signal * correction_factor

    return corrected_signal


# Calcula a amplitude máxima dos dois sinais
max_amplitude = max(np.max(np.abs(resample_100x)), np.max(np.abs(janelamento_012)))

# Aplica correção de amplitude ao sinal com janelamento
corrected_janelado_signal2 = apply_amplitude_correction(janelamento_012, max_amplitude)

plot_fft(resample_100x,target_sample_rate_100,"Resampled 10x")
plot_fft(janelamento_012,target_sample_rate_012,"Resampled 1.2x com janelamento")
plot_fft(corrected_janelado_signal2,target_sample_rate_012,"Resampled 1.2x com janelamento e fator de correção")
```
