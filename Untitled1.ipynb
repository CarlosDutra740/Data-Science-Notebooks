```python
import numpy
E = numpy.array([[-3,4,2,-7],  
              [9,-5,-2,-3],  
              [-6,0,4,8]],)  
print(E)  
 
#L2 <-> L1  
aux = numpy.copy(E[1,:])  
E[1,:] = numpy.copy(E[0,:])  
E[0,:] = numpy.copy(aux)  
print(E)  
 
#zera E[1:2,0]  
E[1,:] = E[1,:] - (E[1,0]/E[0,0])*E[0,:]  
E[2,:] = E[2,:] - (E[2,0]/E[0,0])*E[0,:]  
print(E)  
 
#zera E[2,1]  
E[2,:] = E[2,:] - (E[2,1]/E[1,1])*E[1,:]  
print(E)  
 
#sub. regressiva  
x = numpy.zeros(3)  
x[2] = E[2,3]/E[2,2];  
x[1] = (E[1,3] - E[1,2]*x[2])/E[1,1];  
x[0] = (E[0,3] - E[0,2]*x[2] - E[0,1]*x[1])/E[0,0]  
print(x)
```
