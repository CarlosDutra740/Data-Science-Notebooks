```python
#importar a biblioteca math
import math

#Definir um intervalo [a,b] e um erro

a = 1
b = 1.5
e=0.0001


#Definir função 
def f(x):
	return x**5-6

# teste do teorema de Bolzano
if f(a)*f(b) < 0:
 while (math.fabs(b-a)/2>e):
 	xi=(a+b)/2
 	if (xi) == 0:
 		print("A raiz é :",xi)
 		break
 	else:
 		if f(a)*f(xi)< 0: 
 			b = xi
 		else:
 			a = xi
 print("Valor da raiz É:", xi)

else:
	print("Não há raiz neste intervalo")
 
import numpy as np # Biblioteca para trabalhar fundamentalmente com listas 
import pylab as pl
# ver mais detalhes no seguinte a
X = np.linspace(a, b, 256, endpoint=True)
C= f(X)
pl.plot(X, C)
pl.show()


```
