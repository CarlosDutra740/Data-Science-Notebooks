```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# Carregar os dados
file_path = 'omni2_ikYkUwewB7.lst'
data = pd.read_csv(file_path, delim_whitespace=True, header=None, names=[
    'Year', 'DOY', 'Hour', 'Scalar_B', 'Vector_B_Magnitude', 'SW_Temperature',
    'SW_Proton_Density', 'SW_Plasma_Speed', 'Sunspot_No', 'f10.7_index'])

# Preparar a série temporal
data['Time'] = data['Year'] + (data['DOY'] - 1) / 365
data.set_index('Time', inplace=True)
f107_series = data['f10.7_index']

# Dividir a série em treino e teste (70% treino)
train_size = int(len(f107_series) * 0.7)
train, test = f107_series[:train_size], f107_series[train_size:]

# Criar e comparar modelos com diferentes ciclos sazonais (1 a 12 anos)
results_summary = []

for years in range(1, 13):
    seasonal_period = years * 365  # Convertendo anos para dias

    try:
        # Ajustar o modelo SARIMA com o ciclo sazonal específico
        model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, seasonal_period))
        results = model.fit(disp=False)

        # Adicionar as estatísticas ao resumo
        results_summary.append({
            'Years_Cycle': years,
            'AIC': results.aic,
            'BIC': results.bic
        })

        print(f"Modelo com ciclo de {years} anos: AIC = {results.aic:.2f}, BIC = {results.bic:.2f}")

    except Exception as e:
        print(f"Erro no ajuste do modelo com ciclo de {years} anos: {e}")

# Mostrar o resumo das estatísticas
results_df = pd.DataFrame(results_summary)
print("\nResumo das estatísticas dos modelos:")
print(results_df)

# Plotar os AICs para visualização
plt.figure(figsize=(10, 5))
plt.plot(results_df['Years_Cycle'], results_df['AIC'], marker='o', label='AIC')
plt.plot(results_df['Years_Cycle'], results_df['BIC'], marker='s', label='BIC')
plt.xlabel('Ciclo Sazonal (anos)')
plt.ylabel('Critério de Informação')
plt.title('Comparação de Modelos SARIMA')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeRegressor, export_graphviz
from sklearn import tree
import matplotlib.pyplot as plt
import graphviz
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

# Ler o arquivo com as colunas especificadas
file_path = 'omni2_ikYkUwewB7.lst'
data = pd.read_csv(file_path, delim_whitespace=True, header=None, names=[
    'Year', 'DOY', 'Hour', 'Scalar_B', 'Vector_B_Magnitude', 'SW_Temperature',
    'SW_Proton_Density', 'SW_Plasma_Speed', 'Sunspot_No', 'f10.7_index'])

# Converter o tempo em uma escala contínua de anos
data['Time'] = data['Year'] + (data['DOY'] - 1) / 365

# Selecionar as features e a variável alvo
X = data[['Time', 'Scalar_B', 'Vector_B_Magnitude', 'SW_Temperature',
          'SW_Proton_Density', 'SW_Plasma_Speed', 'Sunspot_No']]
y = data['f10.7_index']  # Sendo 'f10.7_index' a variável alvo

# Dividir o dataset em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar e treinar o modelo de árvore de decisão
model = DecisionTreeRegressor(max_depth=4)
model.fit(X_train, y_train)

# 1. Validação Cruzada
cv_scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')
rmse_scores = np.sqrt(-cv_scores)
print(f'RMSE (validação cruzada): {rmse_scores.mean()} ± {rmse_scores.std()}')

# Fazer previsões
y_pred = model.predict(X_test)

# 2. Avaliação com Métricas
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f'RMSE (teste): {rmse}')
print(f'R²: {r2}')

# 3. Gráfico de Importância das Características
importance = model.feature_importances_
feature_names = X.columns
indices = np.argsort(importance)[::-1]

plt.figure(figsize=(10, 6))
plt.title("Importância das Características")
plt.bar(range(X.shape[1]), importance[indices], align="center")
plt.xticks(range(X.shape[1]), feature_names[indices], rotation=45)
plt.xlim([-1, X.shape[1]])
plt.ylabel("Importância")
plt.show()

# Plotar a árvore de decisão usando o Graphviz
dot_data = tree.export_graphviz(model, out_file=None, feature_names=X.columns, filled=True, rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph.render("decision_tree")  # Salva o gráfico em um arquivo chamado 'decision_tree'

# Exibir a árvore de decisão plotada
graph.view()
```
