```python
# Prof Arian, Aula 2
a1 = a
b1 = b
from numpy import log10 as log
erro = float(input('Informe o valor do erro epsilon '))
k = (log(b1-a1)-log(erro))/log(2) # Estimativa do número de iterações, ver slider 21 da aula 2: (log(b-a)-log(erro))/log(2)< k 
print('Estimativa do números de iterações é k = %i' %k)
i=0
c = b1 - a1 
x0 = (a1 +  b1)/2
while (c >= erro or math.fabs(f(x0)) >= erro):
  if (f(a1)*f(x0)) < 0:    
    b1 = x0
  else:
    a1 = x0  
  x = (a1 +  b1)/2
  c = x - x0
  x0 = x
  i = i + 1
  if i == 100:
    break
ErroRelativo = c / x * 100
print('raiz Xn = %f' %x)   
print('f(Xn) = %f' %f(x)) 
print('iterações n = %i' %(i))
print('|xn - a| = {0} < erro e |f(xn)| = {1} < erro'.format(round(c,5), round(abs(f(x)),5)))
print('|xn - a|/xn = {0} %'.format(round(ErroRelativo,4)))
```
